### Rules of the game & code adaptation

I could not find a precise international version, so I decided to copy and paste the Wikipedia's page, translated from Italian below. For the sake of shortness, only the most important concepts were kept. 

Rag shirt or scamicia is a card game popular in Italy through a deck of 40 divided into 4 suits (cups, coins, sticks, swords) of 10 cards each and played by 2 or more people, even in odd numbers. <br>
The rules of the game are very simple, so much so that it is preferred by children who are starting to learn the use of playing cards. Each player may not use more than one hand during the game. This must be the same both when placing the card and when beating on the deck of cards when 2 identical cards are thrown in a row. <br>

**Arrangement of cards**  <br>
Once the cards have been shuffled and divided by the player on his left, they are divided into equal piles, as many as there are players. Once the turn is over, the player on your right will serve the cards.<br>

The cards must obviously be thrown in different ways depending on the number of players. If there are 2 players they should be gently placed by turning it from the top; if there are 3 players the cards must be turned, always gently, but from the side.<br>

**Game** <br>
Each player holds his own bundle, as he received it and without looking at its contents. The only winning cards are the ace , two and three of any suit.<br>

The first player places the top card of their pile on the table in a visible way. The other players take turns doing the same, so as to form a common pile in the center of the table.<br>

When a player plays one of the three winning cards, the next player is obliged to play the indicated number of cards (1 for the ace, 2 for the two and three for the three). If during these play one of the winning cards comes up, the obligation is interrupted and it is up to the next player to play the number of cards indicated. If the player forced to play the cards does not place a winning card, the pile of cards on the table is taken by the player who played the last card. This will place the cards under yours and start a new game. In Sicily, when a card equal to the previous one is placed, it is necessary to beat the pile with the hand to grab it.<br>

**The endless game** <br>
There is a deck configuration that leads to an infinite game, i.e. in which players will never reach zero cards, as after a certain sequence of moves they will return to the initial state. It was found in 2017 [1] , and is the following (0 indicates a card other than ace, 2 and 3):<br>

Player 1: 0 0 3 0 2 0 2 0 3 0 3 1 0 0 0 0 0 3 1 0 <br>

Player 2: 0 0 1 0 0 2 0 0 0 0 2 0 0 0 0 0 0 0 1 0 <br>

This arrangement has long been unknown, so much so that it has even been classified as an unsolved problem in "Unsolved problems in Combinatorial Games". Small applications can also be found online that can simulate games.<br>

**Score**  <br>
The aim of the game is to take all the cards possible until you remain the only owner. Hence the definition of leaving the other players with their shirts alone . So the score is simply given by the count of the games won."

### _Camin_ interface 

As an approach, I decided to display a more user friendly function which interacts with the user, and in turn calls the actual game function where the decks are exchanged. To be clearer below I wrote a simple schema.

$Camin_{\text{user function}} \Rightarrow Game_{\text{main function}} \Rightarrow Secondary \; functions$

 To make it machine-understandable, cards have been mapped to numbers from 1 to 40, such that each suit is ordered in a dozen (1-10,11-20,21-30,31-40). Some dictionaries will later translate these numbers into cards when needed. Please note that, for each suit, the 8th, 9th and 10th card are neither in numbers nor Jack, Queen, King. They are in fact labeled as Jack Horse and King, but this does not impact the game, since all cards different from Aces, 2s and 3s do not have any impact. 

```python
def camin():
    ###Players creation, interactively###
    p1=input('insert here the name of the first player: ').lower()
    while p1=='':
        print('Please, insert a name, not en empty space')
        p1=input('insert here the name of the first player: ').lower()
    p2=input('insert here the name of the second player: ').lower()
    while p2=='':
        print('Please, insert a name, not en empty space')
        p2=input('insert here the name of the first player: ').lower()
    ##########################################
    ###Decks creation using randomness###
    l=random.sample(range(1,41),40)
    l1=[]
    l2=[]
    for i in range(0,len(l),2):
        l1.append(l[i])
    for i in range(1,len(l),2):
        l2.append(l[i])
    #########################################
    ###Game section###
    ##see game function to understand the returned values##
    start=input('say who is starting: ').lower()
    if start==p1:
        l1,l2,winner=game(l1,l2)
    if start==p2:
        l2,l1,winner=game(l2,l1)
    #########################################
    ###Win check###
    ##checks that no one has won##
    while len(l1)!=0 and len(l2)!=0:
        if winner==False:
            l1,l2,winner=game(l1,l2)
        if winner==True:
            l2,l1,winner=game(l2,l1)
    ###Winner declaration###
    ## print decks to show the final distribution##
    print(printer(l1))
    print(printer(l2))
    
    if len(l1)==0:
        print(f'the winner is {p2},congratulations!')
    else:
        print(f'the winner is {p1}, congratulations!')
    a=input('If you want to play another game write "y" and press enter, otherwise press any key and enter')
    if a=='y':
        camin()
    else:
        return
```

### _Game_ manipulation

After outlining the steps of the game, I wrote down three dictionaries, where each of them creates a link between the mapping of cards in the computer language and the mapping of cards in the human language. More specifically:
**trap**: for each let's say key trap card (Ace, two or three of any suit), it stores as value the number of cards triggered to the other player. <br>
**other**: maps each useless card number to its actual name in the deck. <br>
**trapname**: maps numbers of trap cards to their actual names in the deck. <br>
This is the original reasoning I made, I could have considered optimizing it but I want to underline again the fact that this was an early project of mine. 

```python
trap={1:1,2:2,3:3,11:1,12:2,13:3,21:1,22:2,23:3,31:1,32:2,33:3}
other={4:'four denari',14:'four bastoni',24:'four spade',34:'four coppe',5:'five denari',15:'five bastoni',25:'five spade',35:'five coppe',6:'six denari',16:'six bastoni',26:'six spade',36:'six coppe',7:'seven denari',17:'seven bastoni',27:'seven spade',37:'seven coppe',8:'Jack denari',18:'Jack bastoni',28:'Jack spade',38:'Jack coppe',9:'Horse denari',19:'Horse bastoni',29:'Horse spade',39:'Horse coppe',10:'King denari',20:'King bastoni',30:'King spade',40:'King coppe'}
trapname={1:'ACE denari',2:'two denari',3:' three denari',11:'ACE bastoni',12:'two bastoni',13:'three bastoni',21:'ACE spade',22:'two spade',23:'three spade',31:'ACE coppe',32:'two coppe',33:'three coppe'}
```

The first secondary function introduced is _printer_. It allows for the translation just introduced in the cell above. 

```python
def printer(some):
    ##distinguish between a single element and a list and#
    #translate through the dictionaries from computer language to#
    #human readable language#
    if isinstance(some,int):
        if some in trap:
             return str(trapname[some]).upper()        
    #returns directly the card summoned#
        return str(other[some])
    if isinstance(some,list):
        q=[]
        for x in some:
            if x in trap:
                q.append(str(trapname[x]).upper())
            else:
                q.append(str(other[x]))
    #returns directly the cards summoned#
        return q
```

Given the above information, the main function, where most of the concrete actions are done, can be coded. It will be addressed as "game", and will accept as variables only two lists of cards, not necessarily of the same length. 

```python
def game(l1,l2):    
    ###starting arrangements###
    cards=[] #this is were all played cards are stored#
    for i in range(min(len(l1),len(l2))):
        card1=l1[0]
        print('card extracted from player A: ',printer(card1))
        print(f'deck length:{len(l1)}') #deck length is printed to ensure that the player is
        #aware of his/her situation#
        ###secondary function, see below###
        request()
        ####################################
        card2=l2[0]
        ###decks manipulation###
        l1.remove(card1)
        l2.remove(card2)
        cards.append(card1)
        cards.append(card2)
        ##################################
        ###trap check###
        ## if a card is in the trap dictionary keys then a new trigger function
        ##has to be called##
        if card1 in trap:
            #restore situation where the second card was going to be played#
            cards.remove(card2)
            l2.insert(0,card2)
            ###secondary function, see below###
            cards,winner=move2(l2,l1,trap[card1],cards)
            ##########################################
            break
        #if there are no trap cards from player A then the situation is simpler#
        #going on if B has no traps other cards are extracted, otherwise, #
        #a different function is called.#
        print('card extracted from player B: ',printer(card2))
        print(f'deck length:{len(l2)}')
        ###secondary function, see below###
        request()
        ###################################
        if card2 in trap:   
        ###secondary function, see below###
            cards,winner=move1(l1,l2,trap[card2],cards)
        ###############################################
            break
    ###win (of the whole game) check###
    #enough to check that one of the two has no cards#
    if len(l1)==0 or len(l2)==0:
        winner='finished'
        return l1,l2,winner
    ###########################
   #if there is no game winner, at some point one will anyway win a portion#
   #of the enemy's cards and have to take them from the cards list#
    ###secondary function, see below###
    l1,l2,winner=addcards(cards,winner,l1,l2)
    ##########################################
    #returns the two lists modified and the identity of the winner#
    #if winner is False then the first player holding l1 won, otherwise#
    #the second won. the one that wins starts the following hand.#
    #This can be seen when game is called in the camin function#
    return l1,l2,winner
```

### Secondary functions: _move2, move1, request and addcards_

I coded three manipulation functions and an interactive one. 

_Move1_ and _move2_ are similar but distinct. _Move2_ is involved when the first player extracts a trap card, thus maknig the second player extract a number of cards equal to the correspondent number in the former. _Move1_ does the same but taking into account that the second player triggered this path. I believe that at the time I designed this workflow the reason for this distinction was that first player advantage, which could be simply seen as the fact that he/she plays cards first, makes it more difficult to inglobe the two functions in one. Basically the program should always check the first before the second, and then let the second play, thus the separation of moves. 

```python
def move2(l2,l1,n,cards):
#the first player extracted a card which makes#
#the second extract n cards#
    for i in range(n):
        card2=l2[0]
        print('card extracted from player B: ',printer(card2))
        print(f'deck length:{len(l2)}')
        ###secondary function, see below###
        request()
        ####################################
        ###decks manipulation###
        l2.remove(card2)
        cards.append(card2)
        ####################################
        ###trap check###
        if card2 in trap:
            n=trap[card2]
            #if there is another trap card then the function is flipped#
            move1(l1,l2,n,cards)
            ############################################################
            break
    #if there is no trap card the winner is obviously the first player#
    winner=False
    return cards,winner

def move1(l1,l2,n,cards):
    for i in range(n):
        card1=l1[0]
        print('card extracted from player A: ',printer(card1))
        print(f'deck length:{len(l1)}')
        ###secondary function, see below###
        request()
        ####################################
        ###decks manipulation###
        l1.remove(card1)
        cards.append(card1)
        ####################################
        ###trap check###
        if card1 in trap:
            #if there is another trap card then the function is flipped#
            n=trap[card1]
            move2(l2,l1,n,cards)
            ############################################################
            break
    #if there is no trap card the winner is obviously the second player#
    winner=True
    return cards,winner
```

The _addcards_ function does the work of moving the cards that were won to the round winning player.

```python
def addcards(cards,winner,l1,l2):
#checks winner, prints the outcome of the hand in terms of#
#cards won, and eventually transfers them#
    if winner==True:
        print('player B wins: ',printer(cards))
        request()
        win=l2
    else:
        win=l1
        print('player A wins: ', printer(cards))     
        request()
    for i in range(len(cards)):
        win.append(cards[i])
    #returns the two lists and a boolean value identifying who won#
    return l1,l2,winner
```

_Request_ is an annoying but inevitable function. To allow for interaction with the screen and to make players aware of how the game evolves, for each action request is called, asking if the player wants to continue or exit the game, raising a simple exception which blocks the execution of the whole code. 

```python
def request():
    a=input('press enter or any key+enter to continue, if you want to exit write "exit" and press enter: ')
    if a=='exit':
        raise Exception('The game was terminated')
    return
```
